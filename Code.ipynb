```python
# Date column has been ignored in the dataset as it is not useful for our analysis. Also "rv1" and "rv2" has also been removed

col_temp = ["T1","T2","T3","T4","T5","T6","T7","T8","T9"]

col_hum = ["RH_1","RH_2","RH_3","RH_4","RH_5","RH_6","RH_7","RH_8","RH_9"]

col_weather = ["T_out", "Tdewpoint","RH_out","Press_mm_hg",
                "Windspeed","Visibility"] 
col_light = ["lights"]

col_target = ["Appliances"]
```

```python
# setting the columns for the feature set of variables and the target variable
feature_variables_train = train[col_temp + col_hum + col_weather + col_light ]
target_variables_train = train[col_target]
feature_variables_test = test[col_temp + col_hum + col_weather + col_light ]
target_variables_test = test[col_target]
```

```python
#Using StandardScaler to scale the values of the variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

train_modified = train[list(train_X.columns.values) ]

test_modified = test[list(test_X.columns.values)  ]



sc_train = pd.DataFrame(columns=train_modified.columns , index=train_modified.index)

sc_train[sc_train.columns] = sc.fit_transform(train_modified)

sc_test= pd.DataFrame(columns=test_modified.columns , index=test_modified.index)

sc_test[sc_test.columns] = sc.fit_transform(test_modified)

train_X1=sc_train
test_X1=sc_test
train_y1=train_y
test_y1=test_y
```

```python
import time
from sklearn.svm import SVC

def SupportVectorMachines(kern,train_Xdata,train_ydata,test_Xdata,test_ydata):
    model_SVC = SVC(kernel = kern)
    model_SVC.random_state = 1
    model_SVC_data = {}
    start = time.time()
    model = model_SVC.fit(train_Xdata,train_ydata)
    end = time.time()
    model_SVC_data["Train_Time"] = end - start
    model_SVC_data["Train_Accuracy"] = model_SVC.score(train_Xdata,train_ydata)
    model_SVC_data["Train_Error"] = 1-model_SVC_data["Train_Accuracy"]
    model_SVC_data["Test_Accuracy"] = model_SVC.score(test_Xdata,test_ydata)
    model_SVC_data["Test_Error"] = 1-model_SVC_data["Test_Accuracy"]
    return model_SVC_data,model

```

```python
import pandas as pd
parameters_SVC_linear=[]
parameters_SVC_rbf=[]
parameters_SVC_poly=[]
models_SVM=[]
SVC_Kernels=['linear','rbf','poly']
split=[1,0.9,0.8,0.7,0.6]

for i in SVC_Kernels:
    for j in split:
        trained_SVC=pd.concat([train_X1, train_y1], axis = 1)
        tested_SVC=pd.concat([test_X1,test_y1],axis=1)
        trained_SVC=trained_SVC.sample(frac=j, random_state=1)
        tested_SVC=tested_SVC.sample(frac=j, random_state=1)
        trained_SVC_y=trained_SVC['Appliances']
        del trained_SVC['Appliances']
        trained_SVC_X=trained_SVC
        tested_SVC_y=tested_SVC['Appliances']
        del tested_SVC['Appliances']
        tested_SVC_X=tested_SVC
        model_data_SVC, model_SVM = SupportVectorMachines(i,trained_SVC_X,trained_SVC_y,tested_SVC_X,tested_SVC_y)
        model_data_SVC["Split_Percantage"]=j
        if (i=='linear'):
            parameters_SVC_linear.append(model_data_SVC)
        if (i=='rbf'):
            parameters_SVC_rbf.append(model_data_SVC)
            models_SVM.append(model_SVM)
        if (i=='poly'):
            parameters_SVC_poly.append(model_data_SVC)
            
model_SVM_rbf=models_SVM[0]
```

```python
#Decision Tree implementation for fixing Max-Depth

from sklearn.tree import DecisionTreeClassifier

def DecisionTree1(X_train1,y_train1,X_test1,y_test1,depth):

    model_DT1 = DecisionTreeClassifier(criterion = "entropy",max_depth=depth)
    model_DT1.random_state = 200
    model_DT_data1 = {}
    start = time.time()
    model_DT1.fit(X_train1,y_train1)
    end = time.time()
    model_DT_data1["Train_Time"] = end - start
    model_DT_data1["Train_Accuracy"] = model_DT1.score(X_train1,y_train1)
    model_DT_data1["Train_Error"]= 1-model_DT_data1["Train_Accuracy"]
    model_DT_data1["Test_Accuracy"] = model_DT1.score(X_test1,y_test1)
    model_DT_data1["Test_Error"]= 1-model_DT_data1["Test_Accuracy"]
    return model_DT_data1   
```

```python
#Decision Tree implementation and calculating accuracy of training and test sets after fixing max-depth
from sklearn.tree import DecisionTreeClassifier
import time

def DecisionTree(X_train,y_train,X_test,y_test):
    
    model_DT = DecisionTreeClassifier(criterion = "entropy",max_depth=10)
    model_DT.random_state = 200
    model_DT_data = {}
    start = time.time()
    model_DT.fit(X_train,y_train)
    end = time.time()
    model_DT_data["Train_Time"] = end - start
    model_DT_data["Train_Accuracy"] = model_DT.score(X_train,y_train)
    model_DT_data["Train_Error"]= 1-model_DT_data["Train_Accuracy"]
    model_DT_data["Test_Accuracy"] = model_DT.score(X_test,y_test)
    model_DT_data["Test_Error"]= 1-model_DT_data["Test_Accuracy"]
    return model_DT_data, model_DT  
```

```python
parameters=[]
split=[1,0.9,0.8,0.7,0.6]
print("Decision Tree Classifier (Entropy Criterion)\n")
for i in split:
    trained=pd.concat([train_X, train_y], axis = 1)
    tested=pd.concat([test_X,test_y],axis=1)
    trained=trained.sample(frac=i, random_state=1)
    tested=tested.sample(frac=i, random_state=1)
    trained_y=trained['Appliances']
    del trained['Appliances']
    trained_X=trained
    tested_y=tested['Appliances']
    del tested['Appliances']
    tested_X=tested
    model_data, model_best = DecisionTree(trained_X,trained_y,tested_X,tested_y)
    model_data["Split_Percantage"]=i
    if (i==1):
        Model_Best=model_best
    print('For train data Split Percentage= ',model_data["Split_Percantage"]*100,'\nTraining time=',model_data["Train_Time"],'\nTraning Accuracy= ',model_data["Train_Accuracy"],'\nTesting Accuracy= ',model_data["Test_Accuracy"],'\nTraining Error= ',model_data["Train_Error"],'\nTesting Error= ',model_data["Test_Error"])
    print('\n')
    parameters.append(model_data)
    
    
```

```python
from sklearn.ensemble import GradientBoostingClassifier

def GradientBoosting(X_train2,y_train2,stages,min_sample_lf):
    model_boosting= GradientBoostingClassifier( n_estimators=stages,  min_samples_leaf=min_sample_lf, max_depth=10, random_state=110)
    start=time.time()
    model_boosting=model_boosting.fit(X_train2,y_train2)
    end=time.time()
    Diff=end-start
    return model_boosting,Diff

```

```python
no_stages=[10,20,30,40,50]
parameters_boosting={}
model_boosting_data=[]
for i in no_stages:
    parameters_boosting={}
    model_boostingdata,train_time=GradientBoosting(train_X,train_y,i,1)
    parameters_boosting['Train_Time']=train_time
    parameters_boosting['Train_Accuracy']=model_boostingdata.score(train_X,train_y)
    parameters_boosting['Train_Error']=1-parameters_boosting['Train_Accuracy']
    parameters_boosting['Test_Accuracy']=model_boostingdata.score(test_X,test_y)
    parameters_boosting['Test_Error']=1-parameters_boosting['Test_Accuracy']
    parameters_boosting['Num_Stages']=i
    model_boosting_data.append(parameters_boosting)
    
    
```

```python
TrainingError_boosting=[]
TestingError_boosting=[]
TrainingTime_boosting=[]
for i in model_boosting_data:
    TrainingError_boosting.append(i["Train_Error"])
    TestingError_boosting.append(i["Test_Error"])
    TrainingTime_boosting.append(i["Train_Time"])
    
fig4, ax4 = plt.subplots(nrows=1, ncols=2, figsize=(16,4))

ax4[0].plot(no_stages, TrainingError_boosting, label='Training Set')
ax4[0].plot(no_stages, TestingError_boosting, label='Test Set')
ax4[0].set(xlabel='Number of Stages', ylabel='Error Rate', title='Error rate vs Number of Stages')
ax4[0].grid()
ax4[0].legend(loc='upper right')

ax4[1].plot(TrainingTime_boosting, TrainingError_boosting, label='Training Set')
ax4[1].plot(TrainingTime_boosting, TestingError_boosting, label='Test Set')
ax4[1].set(xlabel='Training Time', ylabel='Error Rate', title='Error Rate vs Training Time')
ax4[1].grid()
ax4[1].legend(loc='upper right')
```
